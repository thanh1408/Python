```python
def get_operation_choice():
    print("Chọn phép tính:")
    print("1. Cộng")
    print("2. Trừ")
    print("3. Nhân")
    print("4. Chia")
    print("5. Chia nguyên")
    print("6. Modulus")
    print("7. Lũy thừa")
    print("8. Thoát")
    choice = input("Vui lòng chọn một phép tính (1-5): ")
    return choice


def get_numbers_from_user():
    try:
        n1 = float(input("Nhập số thứ nhất: "))
        n2 = float(input("Nhập số thứ hai: "))
        return n1, n2
    except ValueError:
        print("Lỗi: Vui lòng nhập số hợp lệ.")
        return get_numbers_from_user()
def add(n1, n2):
    return n1 + n2

def sub(n1, n2):
    return n1 - n2

def mul(n1, n2):
    return n1 * n2

def div(n1, n2):
    if n2 != 0:
        return n1 / n2
    else:
        print("Lỗi: Không thể chia cho 0.")
        return None

def modulus(n1, n2):
    if n2 != 0:
        return n1 % n2
    else:
        return "Lỗi! Chia cho 0."

def chia_nguyen(n1, n2):
    if n2 != 0:
        return n1 // n2
    else:
        return "Lỗi! Chia cho 0."

def luy_thua(n1, n2):
    return n1 ** n2

def main():
    finished = False
    while not finished:
        result = 0
        menu_choice = get_operation_choice()
        
        if menu_choice == '8':
            print("Thoát chương trình.")
            finished = True
            continue
        
        n1, n2 = get_numbers_from_user()
        print(f"Số thứ nhất: {n1}")
        print(f"Số thứ hai: {n2}")


        if menu_choice == '1':
            print("Bạn đã chọn: Cộng")
            result = add(n1, n2)
        elif menu_choice == '2':
            print("Bạn đã chọn: Trừ")
            result = sub(n1, n2)
        elif menu_choice == '3':
            print("Bạn đã chọn: Nhân")
            result = mul(n1, n2)
        elif menu_choice == '4':
            print("Bạn đã chọn: Chia")
            result = div(n1, n2)
        elif menu_choice == '5':
            print("Bạn đã chọn: Chia nguyên")
            result = chia_nguyen(n1, n2)
        elif menu_choice == '6':
            print("Bạn đã chọn: Modulus")
            result = modulus(n1, n2)
        elif menu_choice == '7':
            print("Bạn đã chọn: Lũy thừa")
            result = luy_thua(n1, n2)
        
        if result is not None:
            print('Kết quả:', result)
        print('=================')


main()
```
